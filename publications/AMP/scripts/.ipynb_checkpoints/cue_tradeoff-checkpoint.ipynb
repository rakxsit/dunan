# Cue tradeoff
This script looks at cue tradeoff in acoustic correlates of the geminate contrast in Dunan

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

from scipy.stats import pearsonr
```

# Prep
- Read in files and combine
- Get max and mean f0 and intensity

```python
# Matlab results
df_full = pd.read_csv('../data/AMP_Dunan.csv')

# Original
df_all = pd.read_csv('df_full.csv')
df_closdur = pd.read_csv('df_closdur.csv')
df_vot = pd.read_csv('df_vot.csv')

# UNIQUE ID FROM FILENAME + START OF VOWEL ONSET
df_full['uniqueID'] = df_full['fname'] + '_' + df_full['vowel_t0'].astype(str)

df_all['uniqueID'] = df_all.apply(lambda row: row['Filename'].split('_')[0] + '_' + row['Filename'].split('/')[-1].split('.')[0] + '_' + str(row['vowel_start']), axis = 1)
df_closdur['uniqueID'] = df_closdur.apply(lambda row: row['Filename'].split('_')[0] + '_' + row['Filename'].split('/')[-1].split('.')[0] + '_' + str(row['vowel_start']), axis = 1)
df_vot['uniqueID'] = df_vot.apply(lambda row: row['Filename'].split('_')[0] + '_' + row['Filename'].split('/')[-1].split('.')[0] + '_' + str(row['vowel_start']), axis = 1)
```

```python
# Get vowel duration info into df_full
df_full = df_full.merge(df_all[['uniqueID', 'vdur']], on = 'uniqueID')

# Only the files in df_vot have an identifiable VOT
# Only the files in df_closdur have a reliably measurable closure (i.e. no silence before)
df_full['vot_exists'] = df_full.uniqueID.apply(lambda x: 'yes' if x in df_vot['uniqueID'].unique() else 'no')
df_full['reliable_closure'] = df_full.uniqueID.apply(lambda x: 'yes' if x in df_closdur['uniqueID'].unique() else 'no')
```

```python
df_votexist = df_full.query('vot_exists == "yes"').reset_index(drop = True)
df_reliableclosure = df_full.query('(reliable_closure == "yes")').reset_index(drop = True)
```

```python
# We need VOT and closure duration from df_closdur and df_vot
df_votexist = df_votexist.merge(df_vot[['uniqueID', 'votdur']], on = 'uniqueID')
df_reliableclosure = df_reliableclosure.merge(df_closdur[['uniqueID', 'closdur', 'votdur', 'cdur']], on = 'uniqueID')
```

```python
# SANITY CHECK
print(len(df_full))
print(len(df_votexist))
print(len(df_reliableclosure))
```

```python
df_reliableclosure.groupby('ons').geminate.value_counts()
```

```python
# Get the max and mean measures for f0 and intensity
for df in [df_full, df_votexist, df_reliableclosure]:

    df['max_f0'] = df[[c for c in df if 'F0_' in c]].max(axis=1)
    df['max_int'] = df[[c for c in df if 'Intensity_' in c]].max(axis=1)

    df['mean_f0'] = df[[c for c in df if 'F0_' in c]].mean(axis = 1)
    df['mean_int'] = df[[c for c in df if 'Intensity_' in c]].mean(axis=1)
```

```python
# Get the different duration measures in ms
df_votexist['VOT (ms)'] = df_votexist['votdur'] * 1000
df_votexist['Vowel duration (ms)'] = df_votexist['vdur'] * 1000

df_reliableclosure['Closure duration (ms)'] = df_reliableclosure['closdur'] * 1000
df_reliableclosure['VOT (ms)'] = df_reliableclosure['votdur'] * 1000
df_reliableclosure['Total consonant duration (ms)'] = df_reliableclosure['cdur'] * 1000
df_reliableclosure['Vowel duration (ms)'] = df_reliableclosure['vdur'] * 1000
```

```python
# Create dataframes
df_full[['uniqueID', 'fname', 'subj', 'soundID', 'word', 'ons', 'ons_cat', 'geminate',
         'vowel_t0', 'vowel_t1', 'prev', 'prev_t0', 'prev_t1', 'sentence', 'vdur', 'Fs'] + 
        [c for c in df_full if 'rawF0_' in c] + [c for c in df_full if 'Intensity_' in c]].to_csv('dfAMP/f0-int.csv')

df_votexist[['uniqueID', 'fname', 'subj', 'soundID', 'word', 'ons', 'ons_cat', 'geminate',
         'vowel_t0', 'vowel_t1', 'prev', 'prev_t0', 'prev_t1', 'sentence', 'votdur', 'vdur', 'Fs'] + 
        [c for c in df_full if 'rawF0_' in c] + [c for c in df_full if 'Intensity_' in c]].to_csv('dfAMP/vot.csv')

df_reliableclosure[['uniqueID', 'fname', 'subj', 'soundID', 'word', 'ons', 'ons_cat', 'geminate',
         'vowel_t0', 'vowel_t1', 'prev', 'prev_t0', 'prev_t1', 'sentence', 'closdur', 'cdur', 'votdur', 'vdur', 'Fs'] + 
        [c for c in df_full if 'rawF0_' in c] + [c for c in df_full if 'Intensity_' in c]].to_csv('dfAMP/closure-vot-totalc.csv')
```

## Plot VOT against closure duration

```python
measures = ['Closure duration (ms)', 'VOT (ms)']

print('Pearson R for singleton ', measures[0], 'and', measures[1], 
      pearsonr(x = df_reliableclosure.query('geminate == 0')[measures[0]], 
               y = df_reliableclosure.query('geminate == 0')[measures[1]]))
print('Pearson R for geminate ', measures[0], 'and', measures[1], 
      pearsonr(x = df_reliableclosure.query('geminate == 1')[measures[0]], 
               y = df_reliableclosure.query('geminate == 1')[measures[1]]))

sns.lmplot(data = df_reliableclosure, x = measures[0], y = measures[1], hue = 'geminate', 
           #col = 'geminate', 
           scatter_kws={"s": 20, 'alpha':.5})
```

```python
measures = ['Total consonant duration (ms)', 'VOT (ms)']

print('Pearson R for singleton ', measures[0], 'and', measures[1], 
      pearsonr(x = df_reliableclosure.query('geminate == 0')[measures[0]], 
               y = df_reliableclosure.query('geminate == 0')[measures[1]]))
print('Pearson R for geminate ', measures[0], 'and', measures[1], 
      pearsonr(x = df_reliableclosure.query('geminate == 1')[measures[0]], 
               y = df_reliableclosure.query('geminate == 1')[measures[1]]))

sns.lmplot(data = df_reliableclosure, x = measures[0], y = measures[1], hue = 'geminate', 
           scatter_kws={"s": 20, 'alpha':.5})
```

## Plot f0 against intensity

```python
df_full['mean_int'].min()
```

```python
sns.set(font_scale = 1.5, style = 'whitegrid')
sns.plotting_context(rc={"grid.linewidth": 0.1})

m1, m2 = ('mean_f0', 'mean_int')

print('Pearson R for singleton', m1, 'and', m2, 
      pearsonr(x = df_full.query('geminate == 0')[m1], 
               y = df_full.query('geminate == 0')[m2]))
print('Pearson R for geminate', m1, 'and', m2, 
      pearsonr(x = df_full.query('geminate == 1')[m1], 
               y = df_full.query('geminate == 1')[m2]))

g = sns.lmplot(data = df_full, x = m1, y = m2, hue = 'geminate', hue_order = [1, 0], 
               col = 'geminate', scatter_kws={"s": 20, 'alpha':.5})

#g.fig.suptitle('Dunan')

g.set(xlim = (100, 350), ylim = (0, 70))
g.despine(right = False)

g.fig.subplots_adjust(wspace = 0.12)#, top = 0.87)

for i, ((gem, c), (r, p)) in enumerate(zip([('Singletons', 1), ('Geminates', 0)], [('.26', '< .001***'), ('.46', '< .001***')])):
    
    ax = g.axes.flat[i]
    
    ax.set(title = gem)

    # place a text box in upper right in axes coords
    props = dict(boxstyle='round', facecolor=sns.color_palette()[c], alpha=1)
    ax.text(0.75, 0.97, 'r = ' + r + '\np ' + p, transform=ax.transAxes, fontsize=14,
            verticalalignment='top', bbox=props, color = 'white')

    ax.set(xlabel = 'Mean F0 (Hz)')
    
    if i == 0: 
        ax.set(ylabel = 'Mean Intensity (dB)')

plt.savefig('figAMP/dunan-f0-int.png', dpi = 300, bbox_inches = 'tight', transparent = True)
```
